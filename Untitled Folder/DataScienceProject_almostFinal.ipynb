```python
import json
import requests
import urllib
from urllib.request import urlopen
import pandas as pd
import io
import numpy as np
import sklearn
from sklearn import linear_model, metrics, preprocessing
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Imoprting the results of the 20's Knesset-elections, using API from:
# 'https://data.gov.il/dataset/votes-knesset/resource/929b50c6-f455-4be2-b438-ec6af01421f2'
params = {'resource_id': '929b50c6-f455-4be2-b438-ec6af01421f2',
          'limit': 100000}
url = 'https://data.gov.il/api/3/action/datastore_search'
req = requests.get(url, params=params).json()
votes_df = pd.DataFrame(req['result']['records'])
votes_df
```

```python
elec20_df = votes_df.copy()
```

```python
# Imoprting the results of the 21's Knesset-elections, using API from:
# 'https://votes21.bechirot.gov.il/'
url2 = 'https://media21.bechirot.gov.il/files/expc.csv'
req2 = requests.get(url2).content.decode("ISO-8859-8")
df_votes2 = pd.read_csv(io.StringIO(req2))
df_votes2
```

```python
# Imoprting the cities features:
# 'https://www.cbs.gov.il/he/publications/Pages/2019/%D7%99%D7%99%D7%A9%D7%95%D7%91%D7%99%D7%9D-%D7%91%D7%99%D7%A9%D7%A8%D7%90%D7%9C.aspx'
url3 = "https://www.cbs.gov.il/he/publications/doclib/2019/ishuvim/bycode2020.xlsx"
req = requests.get(url3).content
df3 = pd.read_excel(req)
df3
```

```python
#check for every numeric column which represents a party, if the party passed the votes threshold.
#remove it from the DF if it didn't
cols_check = ['אמת', 'ג',
       'דעם', 'ום', 'ז', 'זי', 'זך', 'זנ', 'זץ', 'טב', 'י', 'יז', 'ין', 'יץ',
       'כ', 'ךק', 'ל', 'מחל', 'מרצ', 'נ', 'נז', 'נך', 'נץ', 'נר', 'ן', 'ןך',
       'ןנ', 'פה', 'ףז', 'ףי', 'ףך', 'ףנ', 'ףץ', 'צק', 'ץ', 'ץז', 'ץי', 'ק',
       'קי', 'קן', 'קף', 'ר', 'שס']
for col in cols_check:
    if(np.sum(elec21_df[col]) < 139000):
        elec21_df.drop(columns=col, inplace=True)
```

```python
#The DF that contains the features of the cities
cities_df = df3.copy()
cities_df
```

```python
#Sorting the DF by the city's number in order to merge it with the elections results DF
sorted_cities_df = cities_df.sort_values(by='סמל יישוב', ascending=True)
sorted_cities_df.columns
```

```python
#Cleaning the data from the cities DF, rows with less than 10 valid items will be romoved and the other NaN will be filled
print(f'sorted_cities_df.shape = {sorted_cities_df.shape}')
print(f'elec21_df.shape = {elec21_df.shape}')
sorted_cities_df.dropna(thresh = 10, inplace=True)
sorted_cities_df.fillna(-1, inplace=True)
print(f'sorted_cities_df.shape AFTER dropna = {sorted_cities_df.shape}')
```

```python
#Merging the DFs. this will be our main DF for the prediction action 
merged_df = sorted_cities_df.merge(elec21_df, right_on='סמל ישוב', left_on='סמל יישוב')
print(f'merged df shape: {merged_df.shape}')
```

```python
#Removing from the DF unusful and duplicated columns
merged_df.drop(columns=['שם ישוב','כשרים', 'סמל ישוב','תעתיק','השתייכות ארגונית','אשכול רשויות מקומיות'], inplace=True)
print(f'merged df shape AFTER drop: {merged_df.shape}')
```

```python
#Fixing a column that not all its values were numeric by turning the non-numeric values to approximated numeric values
merged_df['שנת ייסוד'] = merged_df['שנת ייסוד'].replace({'ותיק':'1880'})
merged_df.shape
```

```python
merged_df_fin = merged_df[['מצביעים','פסולים','מחל','ערבים','סך הכל אוכלוסייה 2020','בזב']]
merged_df_fin['מחוז'] = LabelEncoder().fit_transform(merged_df['מחוז'].copy())
merged_df_fin['אזור טבעי'] = LabelEncoder().fit_transform(merged_df['אזור טבעי'].copy())
merged_df_fin['מעמד מונציפאלי'] = LabelEncoder().fit_transform(merged_df['מעמד מונציפאלי'].copy())
merged_df_fin['דת יישוב'] = LabelEncoder().fit_transform(merged_df['דת יישוב'].copy())
merged_df_fin['צורת יישוב שוטפת'] = LabelEncoder().fit_transform(merged_df['צורת יישוב שוטפת'].copy())
```

```python
merged_df_fin['invalid ratio'] = merged_df_fin['פסולים'] / merged_df_fin['מצביעים']
sns.violinplot(merged_df_fin['דת יישוב'], merged_df_fin['invalid ratio'])
```

```python
#Using relpot I wanted to check the effection of normalization of party votes on the invliad votes.
#Here, without normalization it looks like there is a connection between the likud voters and the invalid votes
#but itmight be just because there are more voters.
#I choosed the Likuk party because it had the largest correlation in the corr() order used earlier.
sns.relplot(x='מחל', y='פסולים', data=merged_df)
```

```python
#And as we can see, noramalization of the data shows there is NO correlation between the pararm
merged_df_fin['invalid ratio'] = merged_df_fin['פסולים'] / merged_df_fin['מצביעים']
df_per = pd.DataFrame()
df_per['likud'] = merged_df_fin['מחל'] / merged_df_fin['מצביעים']
df_per['invalid ratio'] = merged_df_fin['invalid ratio']
sns.relplot(x='likud', y='invalid ratio', data=df_per)
```

```python
#Performing the Linear regression
regressor = LinearRegression()
X = merged_df_fin[['מצביעים','מחוז','מעמד מונציפאלי','צורת יישוב שוטפת','אזור טבעי','דת יישוב','ערבים','סך הכל אוכלוסייה 2020','בזב']]
Y = merged_df_fin['פסולים']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=51)
regressor.fit(X_train, Y_train)
Y_pred = regressor.predict(X_test)
```

```python
#Estimateing the scores of our prediction, first by R2 score 
score = r2_score(Y_test, Y_pred)
print(f'score: {score}')
```

```python
sns.regplot(Y_test, Y_pred)
```

```python
#Put the result of the coef (the correlation between the tested features and the invalid votes) int DF
#We can see that the most efficent feature in the cities religion
res_df = pd.DataFrame(['מצביעים','מחוז','מעמד מונציפאלי','צורת יישוב שוטפת','אזור טבעי','דת יישוב','ערבים','סך הכל אוכלוסייה 2020','בזב'], abs(regressor.coef_),
                       columns = ['Features'])
res_df
```

```python
#Check within the city's relgion the ratio between the 4 religions which are included in the DF
# 1 - Jewish, 2 - non-Jewish, 3 - Bedouin tribe, 4 - Mixed
religions = [1,2,3,4]
invalids = []
voters = []
rate = []
for x in religions:
    invalids.append(merged_df_fin[merged_df_fin['דת יישוב']==x]['פסולים'].sum())
    voters.append(merged_df_fin[merged_df_fin['דת יישוב']==x]['מצביעים'].sum())
    rate.append((invalids[x-1] / voters[x-1])*1000)
res = pd.DataFrame({'מצביעים':voters, 'פסולים': invalids, 'יחס': rate})
res
```

```python
plt_df = pd.DataFrame()
plt_df['religion'] = religions
plt_df['ratio'] = res['יחס']

sns.barplot(x='religion', y='ratio', data=plt_df)
```
